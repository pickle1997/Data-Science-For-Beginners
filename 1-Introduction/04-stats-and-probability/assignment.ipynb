In this assignment, we will use the dataset of diabetes patients taken [from here](https://www4.stat.ncsu.edu/~boos/var.select/diabetes.html).

```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt

data = pd.read_csv("../../data/diabetes.tsv",sep='\t')
data.head()
```


In this dataset, columns as the following:
* Age and sex are self-explanatory
* BMI is body mass index
* BP is average blood pressure
* S1 through S6 are different blood measurements
* Y is the qualitative measure of disease progression over one year

Let's study this dataset using methods of probability and statistics.

### Task 1: Compute mean values and variance for all values

```python
data
mean_all = data[['AGE','SEX','BMI','BP','S1','S2','S3','S4','S5','S6','Y']].mean()
var_all = data[['AGE','SEX','BMI','BP','S1','S2','S3','S4','S5','S6','Y']].var()
std_all = data[['AGE','SEX','BMI','BP','S1','S2','S3','S4','S5','S6','Y']].std()
print(mean_all, var_all, std_all)
```

### Task 2: Plot boxplots for BMI, BP and Y depending on gender

```python
data.boxplot(column='BMI', by='SEX', figsize=(10,8))
plt.xticks(rotation='vertical')
plt.tight_layout()
plt.ylabel("BMI")
plt.show()
```

```python
data.boxplot(column='BP', by='SEX', figsize=(10,8))
plt.xticks(rotation='vertical')
plt.tight_layout()
plt.ylabel("Blood Pressue (mmHg)")
plt.show()
```

```python
data.boxplot(column='Y', by='SEX', figsize=(10,8))
plt.xticks(rotation='vertical')
plt.tight_layout()
plt.ylabel("Disease Progression")
plt.show()
```
